```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
df=pd.read_csv("CHURN.csv")
df.head()
```

```python
df.drop('customerID',axis=1,inplace=True)
df.head()
```

```python
df.dtypes
```

```python
df=df[df['TotalCharges']!=" "]
```

```python
df[df['TotalCharges']==" "]
```

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'])
df.dtypes
```

```python
df2=df
```

```python
for col in df2:
    if df2[col].dtypes=='object':
        print(f'{col} = {df2[col].unique()}')
```

```python
df2=df2.replace({"Yes":1,"No":0})
```

```python
df2=df2.replace({"No phone service":0})
for col in df2:
    if df2[col].dtypes=='object':
        print(f'{col} = {df2[col].unique()}')
```

```python
df2=df2.replace({"No internet service":0})
for col in df2:
    if df2[col].dtypes=='object':
        print(f'{col} = {df2[col].unique()}')
```

```python
dic={'InternetService':{'DSL':1, 'Fiber optic':2},'gender' :{'Female':0, 'Male':1},'Contract': {'Month-to-month':0 ,'One year':1 , 'Two year':2},
        'PaymentMethod': {'Electronic check':0, 'Mailed check':1, 'Bank transfer (automatic)':2,'Credit card (automatic)':3}}
df2=df2.replace(dic)
for col in df2:
        print(f'{col} = {df2[col].unique()}')
```

```python
df2.TotalCharges.describe()
```

```python
df2.tenure =df2.tenure/72.000000
df2.MonthlyCharges=df2.MonthlyCharges/118.750000
df2.TotalCharges=df2.TotalCharges/8684.800000
df2.head()
```

```python
from sklearn.model_selection import train_test_split
X=df2.drop('Churn',axis=1)
y=df2.Churn
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
X_train
```

```python
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(19,),activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train,y_train,epochs=100)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_predicted=[]
for i in y_pred:
    if i<0.5:
        y_predicted.append(0)
    elif i>=0.5:
        y_predicted.append(1)

y_predicted
```
